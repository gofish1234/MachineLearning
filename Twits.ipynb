```python
from textblob import TextBlob
import pandas as pd
```

```python
df = pd.read_csv('../Resources/AMZN_WMT_TGT_HD_LOW_M_BBY_twits.csv', encoding = "ISO-8859-1")
df.head()
```

```python
df = df.astype('str', copy=False)
```

```python
df[['Twitpolarity', 'Twitsubjectivity']] = df['Twit'].apply(lambda Text: pd.Series(TextBlob(Text).sentiment))
df.head(100)
```

```python
Group_df = df.groupby(['Date', 'Stock'])['Twitpolarity', 'Twitsubjectivity'].mean()
Group_df.to_csv("Twit_Polarity_Subjectivity_mean.csv", index=False, header=True)
Group_df
```
