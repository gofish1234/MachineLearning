```python
from textblob import TextBlob
import pandas as pd
```

```python
df = pd.read_csv('stock_tweets_all.csv', encoding = "ISO-8859-1")
df.head()
```

```python
df = df.astype('str', copy=False)
```

```python
df[['twitpolarity', 'twitsubjectivity']] = df['tweet'].apply(lambda Text: pd.Series(TextBlob(Text).sentiment))
df.head(100)
```

```python
Group_df = df.groupby(['date', 'stock'])['twitpolarity', 'twitsubjectivity'].mean()
Group_df.to_csv("polar_subject_mean.csv", index=False, header=True)
Group_df
```
